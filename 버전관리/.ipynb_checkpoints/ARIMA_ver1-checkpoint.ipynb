```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.stattools import adfuller
import statsmodels.api as sm
```

```python
tm1 = tm.groupby(['일시']).mean().reset_index()
tm1 = tm1[['일시', '가격']]
tm2 = tm1.set_index('일시')
```

```python
tm2
```

```python
tm2.iloc[1825]
```

```python
tm_cut = tm2.iloc[:1826]
```

```python

tm_cut.plot(figsize=(15,4))
```

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(tm_cut)
plot_pacf(tm_cut)
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

diff_1=tm_cut.diff(periods=1).iloc[1:] # 차분
diff_1.plot(figsize=(15,4))
plot_acf(diff_1)
plot_pacf(diff_1)
plt.show()
```

```python
model = ARIMA(tm_cut, order=(0,1,0))
model_fit = model.fit(trend='c', full_output=True, disp=1)
print(model_fit.summary())
```

```python
model_fit.plot_predict()
```

```python
model_fit.forecast(steps=1)
```
