```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.stattools import adfuller
import statsmodels.api as sm
```

```python
tm1 = tm.groupby(['일시'])['가격'].mean()
tm1 = pd.DataFrame(tm1)
tm2 = tm1.reset_index()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(tm1, freq=365)  
fig = plt.figure()  
fig = decomposition.plot()
```

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(tm1):

    #Determing rolling statistics
    rolmean = tm1.rolling(12).mean()
    rolstd = tm1.rolling(12).std()

    #Plot rolling statistics:
    fig = plt.figure(figsize=(10, 6))
    orig = plt.plot(tm1, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')

    plt.legend(loc='best'); plt.title('Rolling Mean & Standard Deviation')
    plt.show()

    #Perform Dickey-Fuller test:
    print ('<Results of Dickey-Fuller Test>')
    dftest = adfuller(tm1, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4],
                         index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
```

```python
test_stationarity(tm1['가격'])
```

```python
tm1['first_difference'] = tm1['가격'] - tm1['가격'].shift(1)  
# Or Alternatively,
# df.diff().plot()
test_stationarity(tm1.first_difference.dropna(inplace=False))
```

```python
tm1['seasonal_first_difference'] = tm1['first_difference'] - tm1['first_difference'].shift(12)  
test_stationarity(tm2.seasonal_first_difference.dropna(inplace=False))
```

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(tm2.seasonal_first_difference.iloc[13:], lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(tm2.seasonal_first_difference.iloc[13:],lags=40,ax=ax2)
```

```python
mod = sm.tsa.SARIMAX(tm2['가격'], trend='ct',order=(1,2,1), seasonal_order=(1,1,1,12))
results = mod.fit()
print (results.summary())
```

```python
tm2['forecast'] = results.predict(start = len(tm2)-12, end= len(tm2), dynamic= True)  
tm2[['가격', 'forecast']].plot()
tm2[-12:]
```
